```python
import pandas as pd
```

```python
import numpy as np
```

```python
path = "/content/drive/MyDrive/Data mining/adult.csv"
```

```python
df = pd.read_csv(path)
df.head(2)
```

```python
x = pd.pivot_table(df,                     
               index= ['marital-status', 'education'],
               aggfunc = 'median',
               values = 'age',
               columns = 'sex',
               margins = True        
               )  
x
```

```python
y = pd.pivot_table(df,                     
               index= ['marital-status', 'education'],
               aggfunc = ['mean','median'],
               values = 'age',
               margins = True  ,
               columns = 'sex'   
               )  
y
```

```python
 # 2 Generate a pivot table showing the number of people for each marital status.
 z = pd.pivot_table(df,                     
               index= 'marital-status',
               aggfunc = 'count',
               values = 'age',
               margins = True  ,
               columns = 'sex'   
               )  
z
```
