```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Calculate correlation coefficient
corr_coef = np.corrcoef(x, y)[0, 1]

# Create scatterplot
plt.scatter(x, y)
plt.title(f"Correlation Coefficient: {corr_coef:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
#In this example, we use the NumPy library to generate two sets of 100 random numbers drawn from a normal distribution with mean 0 and standard deviation 1. We then calculate the correlation coefficient using the np.corrcoef() function and plot the data using plt.scatter(). Since the two sets of data have little to no linear relationship, the resulting correlation coefficient will be close to 0.
#The resulting scatterplot should look like a cloud of points with no clear pattern or trend, indicating little to no linear relationship between the variables. The correlation coefficient should also be close to 0, which confirms the lack of linear relationship
```

```python
import pandas as pd
import statsmodels.api as sm

# load data
path = 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/census.csv'
data = pd.read_csv(path)
data
```

```python
# perform linear regression analysis
X = data['NETMIG2012']
y = data['STATE']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# calculate residuals
resid = model.resid

# calculate median value of residuals
median_resid = resid.median()

# calculate interquartile range (IQR) of residuals
q1 = resid.quantile(0.25)
q3 = resid.quantile(0.75)
iqr = q3 - q1

print('Median value of residuals: {:.4f}'.format(median_resid))
print('Interquartile range (IQR) of residuals: {:.4f}'.format(iqr))
```
