```python
import pandas as pd
```

```python
import numpy as np
```

```python
path2 = "/content/drive/MyDrive/Data mining/antropometria-dataset-2.csv"
```

```python
df = pd.read_csv(path2)
df
```

```python
# 3. Generate a pivot table showing the average weight of people by state.
x = pd.pivot_table(df,                     
               index= 'desc_ent',
               aggfunc = 'mean',
               values = 'peso',
               margins = True       
               )  
x
```

```python
# 4. Add columns showing the average weight by gender by state.
y = pd.pivot_table(df,                     
               index= 'desc_ent',
               aggfunc = 'mean',
               values = 'peso',
               columns = 'sexo',
               margins = True       
               )  
y
```
