```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/tec03/Datasets/main/datasets/satis.csv')
mpl.style.use('ggplot')  # optional: for ggplot-like style
```

```python
df.head(5)
```

Let's check how many missing values are there in each column:

```python
df.isnull().sum()
```

Let's check how many missing values are there in each Row:

```python
df.dropna(axis = 1, how = 'all', inplace = True)
df.dropna(how = 'all', inplace = True)
```

```python
df.isnull().sum()
```

Let us drop the column reason, since 18 out of 50 data is missing:

```python
df.drop('reason', 
        axis = 1, 
        inplace = True)
df.head(2)
```

```python
df.isnull().sum()
```

```python
df['Age']
```

```python
df['Age'] = df['Age'].astype(int)
df.Age
```

```python
df.head(2)
```

```python
print(df.Age.min())
print(df.Age.max())
```

```python
def custom_cls(x):
  if(0<= x <= 19):
    return '<20'
  elif (20 <= x <= 29):
    return '20 - 29'
  elif (30 <= x <= 39):
    return '30 - 39'
  return '>=40'
```

```python
df.head(2)
```

```python
print(custom_cls(0))
print(custom_cls(10))
print(custom_cls(20))
print(custom_cls(30))
print(custom_cls(40))
print(custom_cls(-5))
```

```python
df['AgeGP'] = df.apply(lambda tbl: custom_cls(tbl['Age']),
                       axis = 1)
df.head(10)
```

```python
df['AgeGP'].value_counts()
```

```python
counts = df['AgeGP'].value_counts()
type(counts)
```

```python
counts = counts.to_frame()
```

```python
counts
```

```python
suma = counts.iloc[ : , : ].sum()
suma
```

```python
counts['%'] = (counts.iloc[ : , : ]/suma)*100
counts
```

```python
counts.index
```

```python
counts.reindex(['<20', '20 - 29', '30 - 39', '>=40'])
```

```python
counts.reset_index(inplace = True)
```

```python
counts
```

```python
counts.rename(columns = {'index':'AgeGP',
               'AgeGP':'Counts'},
              # axis = 1,   o se usa columns =  o se usa axis = 1
              inplace = True)
counts 
```

Now, we make a pie chart.

Each slice represent the size of the corresponding value in the column
Defalut label is index
we can specify any column as label

```python
colors_fav = ['gold','yellowgreen','coral','skyblue','lavender','darkviolet']
space_fav = [ 0.02 , 0.03 , 0.04 , 0.05 ]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
counts['Counts'].plot(kind = 'pie',
                      labels = None,
                      figsize = (15,6),
                      autopct = '%1.3f%%', #1.3 nos da 3 decimales, f= float
                      pctdistance = 1.2,
                      #shadow = True
                      colors = colors_fav,
                      startangle = 90,
                      explode = space_fav   #para separar las rebanadas
                      );
plt.legend(labels = counts.AgeGP,
           bbox_to_anchor = (1,1)
           );
```

```python
plt.style.use('ggplot')
```

```python
x = counts.AgeGP
y = counts.Counts
plt.bar(x,y)
```

```python
plt.subplot(1,2,1)  #no es necesario poner comas , se puede poner 121
counts['Counts'].plot(kind = 'pie',
                      labels = None,
                      figsize = (15,6),
                      autopct = '%1.3f%%', #1.3 nos da 3 decimales, f= float
                      pctdistance = 1.2,
                      #shadow = True
                      colors = colors_fav,
                      startangle = 90,
                      explode = space_fav   #para separar las rebanadas
                      );
plt.legend(labels = counts.AgeGP,
           bbox_to_anchor = (1,1)
           );

plt.subplot(1,2,2)      
x = counts.AgeGP
y = counts.Counts
plt.bar(x,y, color = 'navy')
plt.savefig('sample_data/comp.png')
```
