```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px   #Graficas interactivas
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/antropometria-dataset-1.csv',
                   #sheet_name='Canada by Citizenship',
                   #skiprows=range(20),
                   #skipfooter=2,
                   #engine='openpyxl'
                  )
df.head(2)
```

```python
df.columns
```

```python
df_pivot = df.pivot_table('peso', 
               index='desc_ent', 
               columns='sexo'
               )
df_pivot
```

```python
favs = ['folio', 'intp', 'entidad', 'edad', 'meses',
       'ropa', 'emb', 'temb', 'cintura', 'peso', 'sistol',
       'diastol', 'hpresion', 'tbrazo', 'htension', 'PrimaryLast', 'code_upm',
       'est_dis', 'est_urb', 'est_marg', 'pondef', 'est_var', 'talla']

df.drop(favs, 
        axis=1, #axis One is Column
        inplace=True
       )

df.head(6)
```

```python
df2 = df.drop(df[df['sexo']==2].index)
df2
```

```python
df3 = df2.drop(['sexo'], 
                axis = 1)
df3
```

```python
df3_pivot = df3.pivot_table('cadera', 
               index='desc_ent',
               aggfunc='median' 
               )
df3_pivot
```

```python
df3_pivot.plot(kind='line', 
              figsize=(8, 6)
             )

plt.title('Box plot')
plt.ylabel('Y')

plt.show()
```
