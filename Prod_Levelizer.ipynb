```python
import pandas as pd
```

```python
path = "/content/sample_data/Prod. Levelizer database.xlsx"
```

```python
df = pd.read_excel(path)
```

```python
# Reemplazar NaN con 0 en todo el DataFrame
df.fillna(0, inplace=True)
df['GAP Entradas-Salidas'] = df['Entradas (Producción)'] - df['Salidas (Ordenes)']
# Calcular el GAP Acumulado por 'Part Number'
df['GAP Acumulado'] = df.groupby('Part Number')['GAP Entradas-Salidas'].cumsum()
df['Inventario Forecast'] = df['Most Recent Current Inventory']+df['GAP Acumulado']
```

```python
df
```

```python
def calcular_porcentaje_adj (inventario_forecast, minimo_inventario, maximo_inventario):
    if inventario_forecast < minimo_inventario:
        return 1.10  # 110%
    elif inventario_forecast > maximo_inventario:
        return 0.90  # 90%
    else:
        return 1.00  # 100%

# Aplicar la función a cada fila del DataFrame
df['Production % Adj.'] = df.apply(lambda row: calcular_porcentaje_adj(row['Inventario Forecast'], row['Minimo Inventario'], row['Maximo Inventario']), axis=1)
```

```python
df
```

```python
df['Entradas (Producción) 1'] = df['Entradas (Producción)'] * df['Production % Adj.']
df['GAP Entradas-Salidas 1'] = df['Entradas (Producción) 1'] - df['Salidas (Ordenes)']
# Calcular el GAP Acumulado por 'Part Number'
df['GAP Acumulado 1'] = df.groupby('Part Number')['GAP Entradas-Salidas 1'].cumsum()
df['Inventario Forecast 1'] = df['Most Recent Current Inventory']+df['GAP Acumulado 1']
```
