# **Eliminar outliers usando IQR**

Detectar outlier es tedioso, especialmente cuando se tienen multiples tipos de datos.

Por lo tanto, tenemos diferentes formas de detectar valores at√≠picos para diferentes tipos de datos.

En cuanto a los datos distribuidos normalmente, podemos obtener el m√©todo Z-Score;

Para skewed data, se usa IQR.

# **IQR es la diferencia entre el cuartil 75th and 25th.**

```python
def remove_outliers_iqr(df2_upsampled, columns):
    for col in columns:
        q1 = df2_upsampled[col].quantile(0.25)
        q3 = df2_upsampled[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df2_upsampled = df2_upsampled[(df2_upsampled[col] >= lower_bound) & (df2_upsampled[col] <= upper_bound)]
    return df2_upsampled

# Se√±ale las columnas para remover los outliers
columns_to_check = [
    "loan_status",
    "loan_amnt",
    "int_rate",
    "dti",
    "fico_range_low",
    "fico_range_high",
    "term",
    "tot_hi_cred_lim",
    "total_acc",
    "total_bal_ex_mort",
    "total_bc_limit",
    "total_rec_late_fee",
    "percent_bc_gt_75",
    "avg_cur_bal",
    "total_il_high_credit_limit"
]

# Solicitar la funci√≥n que remueve los outliers usando IQR
df2_clean = remove_outliers_iqr(df2_upsampled, columns_to_check)

# Mostrar el resultado en el dataframe
df2_clean.head()
```

# **Algoritmo de ensamble (Boosting) XGBoost**
Boosting es un enfoque de Machine Learning basado en la idea de crear una regla de predicci√≥n altamente precisa combinando muchas reglas relativamente d√©biles e imprecisas. El Boosting asume la disponibilidad de un algoritmo de aprendizaje base o d√©bil que, dado ejemplos de entrenamiento etiquetados, produce un clasificador base o d√©bil. La idea fundamental detr√°s de Boosting es elegir conjuntos de entrenamiento para el algoritmo de aprendizaje base de tal manera que lo obligue a inferir algo nuevo sobre los datos cada vez que se lo ejecute.
