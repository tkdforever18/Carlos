```python
import pandas as pd
import statsmodels.api as sm

# load data
path = 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/census.csv'
data = pd.read_csv(path)
data
```

```python
# perform linear regression analysis
X = data['NETMIG2012']
y = data['STATE']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# calculate residuals
resid = model.resid

# calculate median value of residuals
median_resid = resid.median()

# calculate interquartile range (IQR) of residuals
q1 = resid.quantile(0.25)
q3 = resid.quantile(0.75)
iqr = q3 - q1

print('Median value of residuals: {:.4f}'.format(median_resid))
print('Interquartile range (IQR) of residuals: {:.4f}'.format(iqr))
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the data
path = 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/census.csv'
df = pd.read_csv(path)

# Fit a linear regression model
model = sm.OLS.from_formula('NETMIG2012 ~ STATE', data=df)
result = model.fit()

# Get the predicted values
pred = result.predict()

# Calculate the median and IQR of the predicted values
median_pred = pred.median()
iqr_pred = pred.quantile(0.75) - pred.quantile(0.25)

print("Median of the predicted values:", median_pred)
print("IQR of the predicted values:", iqr_pred)
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model

# Load the dataset
path = 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/census.csv'
df = pd.read_csv(path)

# Drop rows with missing values
ndf1 = df.dropna(subset=['NETMIG2012', 'STATE'])

# Define the features (X) and target variable (y)
X = ndf1[['STATE']]
y = ndf1['NETMIG2012']

# Fit the linear regression model
lm = linear_model.LinearRegression()
model_lm = lm.fit(X, y)

# Make predictions for all the states
states = np.array(df['STATE'].unique()).reshape(-1, 1)
pred = model_lm.predict(states)

# Calculate the median and IQR of the predicted values
median_pred = np.median(pred)
iqr_pred = np.percentile(pred, 75) - np.percentile(pred, 25)

print("The median value of the predicted value is:", median_pred)
print("The Inter Quartile Range (IQR) of the predicted value is:", iqr_pred)
```
